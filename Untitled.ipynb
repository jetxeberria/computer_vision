```python
from __future__ import print_function
import SimpleITK as sitk
import sys, os
from matplotlib import pyplot as plt
import skimage
plt.rcParams["image.cmap"]='gray'
```

```python
input_fname = "datos/png/I3T/IMG0001.png"
reader = sitk.ImageFileReader()
reader.SetFileName(input_fname)
fname = reader.GetFileName()
#print ("Input_fname: ", input_fname)
#print ("Reader: ", reader)
#print ("fname: ", fname)
image = reader.Execute()
#print ("Image: ", image)
#print ("Image data type: ", type(image))
#sitk.Show(image)
```

```python
# Simple smoothing
smooth = sitk.SmoothingRecursiveGaussian ( image, 2.0)
smooth_size = smooth.GetSize()
image_size = image.GetSize()
print ("Smooth dtype {}, shape {}".format(type(smooth), smooth_size))
print ("Image dtype {}, shape {}".format(type(image), image_size))
print ("Image. \n\tPixel id type: {}\n\tDimensions: {}".format(image.GetPixelIDTypeAsString(), image.GetDimension()))
print ("Smooth. \n\tBefore. \n\t\tPixel id type: {}\n\t\tDimensions: {}".format(smooth.GetPixelIDTypeAsString(), smooth.GetDimension()))
smooth_cast = sitk.Cast ( smooth, image.GetPixelIDValue())
print ("\tAfter. \n\t\tPixel id type: {}\n\t\tDimensions: {}".format(smooth_cast.GetPixelIDTypeAsString(), smooth_cast.GetDimension()))
print ("Smooth dtype {}, shape {}".format(type(smooth_cast), smooth_cast.GetSize()))
#sitk.Show(sitk.Subtract ( image ,smooth_cast), "DiffWithGaussian")
```

```python
import numpy as np
print ("smooth_sitk_image dtype: \n  {}, dshape: {}, pixeldtype: {}".format(type(smooth_cast), smooth_cast.GetSize(), smooth_cast.GetPixelIDTypeAsString()))
print ("image_sitk_image dtype: \n  {}, dshape: {}, pixeldtype: {}".format(type(image), image.GetSize(),image.GetPixelIDTypeAsString()))
sub_img_smo_cas = sitk.Subtract(image, smooth_cast)
print ("sub_img_smo_cas_sitk_image dtype: \n  {}, dshape: {}, pixeldtype: {}".format(type(sub_img_smo_cas), sub_img_smo_cas.GetSize(),sub_img_smo_cas.GetPixelIDTypeAsString()))
#sitk.Show(sub_img_smo_cas, "Subtract: image - smooth_cast")
image_asarray = sitk.GetArrayFromImage(image)
smooth_cast_asarray = sitk.GetArrayFromImage(smooth_cast)
sub_img_smo_cas_asarray = sitk.GetArrayFromImage(sub_img_smo_cas)
print ("number of components per pixel: ", image.GetNumberOfComponentsPerPixel())
print ("Pixel vector: ", image.GetPixel(0,0,0))

fig, axes = plt.subplots(2,2, figsize=(15,15))
ax00 = axes[0, 0]
ax01 = axes[0, 1]
ax10 = axes[1, 0]
ax11 = axes[1, 1]

print ("image_asarray dtype: \n\t{}, dshape: {}, pixeldtype: {}".format(type(image_asarray), image_asarray.shape, image_asarray.dtype))
ax00.imshow(image_asarray)
ax00.set_title("Image 1")

print ("smooth_cast_asarray dtype: \n\t{}, dshape: {}, pixeldtype: {}".format(type(smooth_cast_asarray), smooth_cast_asarray.shape, smooth_cast_asarray.dtype))
ax01.imshow(smooth_cast_asarray)
ax01.set_title("Smooth")

print ("sub_img_smo_cas_asarray dtype: \n\t{}, dshape: {}, pixeldtype: {}".format(type(sub_img_smo_cas_asarray), sub_img_smo_cas_asarray.shape, sub_img_smo_cas_asarray.dtype))
ax10.imshow(sub_img_smo_cas_asarray)
ax10.set_title("Image - smooth_cast")

ax11.imshow(smooth_cast_asarray[...,0])
ax11.set_title("Smooth_0")

plt.show()

print ("image_asarray:\n{}".format(image_asarray))
print ("smooth_cast_asarray:\n{}".format(smooth_cast_asarray))
print ("sub_img_smo_cas_asarray:\n{}".format(sub_img_smo_cas_asarray))
print (np.setdiff1d(smooth_cast_asarray[...,3], smooth_cast_asarray[...,1]))
print (smooth_cast_asarray[..., 0].all() == smooth_cast_asarray[..., 1].all())
```

```python
image_asarray = sitk.GetArrayViewFromImage(image)
smooth_cast_asarray = sitk.GetArrayViewFromImage(smooth_cast)
sub_img_smo_cas_asarray = sitk.GetArrayViewFromImage(sub_img_smo_cas)

fig, axes = plt.subplots(2,2, figsize=(15,15))
ax00 = axes[0, 0]
ax01 = axes[0, 1]
ax10 = axes[1, 0]
ax11 = axes[1, 1]

print ("image_asarray dtype: {}, dshape: {}, pixeldtype: {}".format(type(image_asarray), image_asarray.shape, image_asarray.dtype))
ax00.imshow(image_asarray)
ax00.set_title("Image 1")

print ("smooth_cast_asarray dtype: {}, dshape: {}, pixeldtype: {}".format(type(smooth_cast_asarray), smooth_cast_asarray.shape, smooth_cast_asarray.dtype))
ax01.imshow(smooth_cast_asarray)
ax01.set_title("Smooth")

print ("sub_img_smo_cas_asarray dtype: {}, dshape: {}, pixeldtype: {}".format(type(sub_img_smo_cas_asarray), sub_img_smo_cas_asarray.shape, sub_img_smo_cas_asarray.dtype))
ax10.imshow(sub_img_smo_cas_asarray)
ax10.set_title("Image - smooth_cast")

ax11.imshow(smooth_cast_asarray[...,0])
ax11.set_title("Smooth")

plt.show()
```
